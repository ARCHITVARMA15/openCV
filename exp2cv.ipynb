```python

b,g,r = cv2.split(img)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")


plt.subplot(1,4,2)
plt.imshow(r,cmap="Reds")
plt.title("Red Layer")
plt.axis("off")


plt.subplot(1,4,3)
plt.imshow(g,cmap="Greens")
plt.title("Green Layer")
plt.axis("off")


plt.subplot(1,4,4)
plt.imshow(b,cmap="Blues")
plt.title("Blue Layer")
plt.axis("off")
```

```python
gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
plt.imshow(gray ,  cmap='gray')
plt.axis("off")
plt.show()
```

```python
#Convert to negative using the formula S = (L-1)-r where L-1 is 255 
import numpy as np 
height, width = gray.shape
negative_img = np.zeros_like(gray)
#loop 
for y in range(height):
    for x in range(width):
        negative_img[y, x] = 255 - int(gray[y, x])


plt.imshow(negative_img, cmap='gray' )
plt.axis("off")
plt.show()
```

```python
#Thresholding using if else 
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
threshold_value = 127 

height, width = gray.shape
thresh = np.zeros_like(gray)

for y in range(height):
    for x in range(width):
        #checking 
        if int(gray[y,x]) > threshold_value:
            thresh[y,x] = 255 #white 
        else:
            thresh[y,x] = 0 # black

plt.figure(figsize=(10,4))

plt.subplot(1, 2, 2)
plt.imshow(thresh, cmap="gray")
plt.title(f"Threshold image (Threshold = {threshold_value})")
plt.axis("off")

plt.show()
```

```python
#Thresholding using if else 
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
threshold_value = 80 

height, width = gray.shape
thresh = np.zeros_like(gray)

for y in range(height):
    for x in range(width):
        #checking 
        if int(gray[y,x]) > threshold_value:
            thresh[y,x] = 255 #white 
        else:
            thresh[y,x] = 0 # black

plt.figure(figsize=(10,4))

plt.subplot(1, 2, 2)
plt.imshow(thresh, cmap="gray")
plt.title(f"Threshold image (Threshold = {threshold_value})")
plt.axis("off")

plt.show()
```

```python
#contrast 

min_val = np.min(gray)
max_val = np.max(gray)

contrast = ((gray - min_val) / (max_val - min_val) * 255).astype(np.uint8)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray, cmap="gray")
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(contrast, cmap="gray")
plt.title("Contrast Stretching")
plt.axis("off")

plt.show()
```

```python
#Contrast Strectching 

contrast = cv2.normalize(gray, None, 0, 255, cv2.NORM_MINMAX)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray , cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(contrast  , cmap='gray')
plt.title("Contrast")
plt.axis("off")

plt.show()
```

```python
#Gray level slicimg without backgorund 
import numpy as np
import matplotlib.pyplot as plt


lower = 80
upper = 170
height, width = gray.shape


slice_img = np.zeros_like(gray)


for y in range(height):
    for x in range(width):
        
        if int(gray[y, x]) >= lower and int(gray[y, x]) <= upper:
            slice_img[y, x] = 255   
        else:
            slice_img[y, x] = 0     


plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(slice_img, cmap='gray')
plt.title("Gray Level Slicing without Background")
plt.axis("off")

plt.show()
```

```python
#gray Level Slicing with Background 
import numpy as np
import matplotlib.pyplot as plt


lower = 80
upper = 170
height, width = gray.shape


slice_img = np.zeros_like(gray)


for y in range(height):
    for x in range(width):
        
        if int(gray[y, x]) >= lower and int(gray[y, x]) <= upper:
            slice_img[y, x] = 255       
        else:
            slice_img[y, x] = gray[y, x]  #prserve the background 

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(slice_img, cmap='gray')
plt.title("Gray Level Slicing (with background)")
plt.axis("off")

plt.show()

```

```python
#Contrast Strecthing 

def contrast(img, a, b):
    l=0.7
    n = 0.4
    m=2
    contrasted = gray.copy()
    v = l*a
    w= m*(b-a) +v
    for x in range(height):
        for y in range(width):
            r = gray[x,y]
            if gray[x,y] >=0 and gray[x,y]<a:
                contrasted[x,y] = 1*r
            elif r>=a and r<b:
                contrasted[x,y] = m*(r-a)+v
            else:
                contrasted[x,y] = n*(r-b) +w
    return contrasted


```

```python
contrasted_img = contrast(img, 70,100)
plt.imshow(contrasted_img, cmap='gray')
plt.axis("off")
plt.show()
```
